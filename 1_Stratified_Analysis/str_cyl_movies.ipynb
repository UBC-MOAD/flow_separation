```python
def gen_anim(_dir,doutcase,maxoutcase,fname):
    fig, ax = plt.subplots(figsize=(10,10))

    ax.set_xlim([-0.1, 0.3])
    ax.set_ylim([0, 0.4])

    XC = np.squeeze(rdmds(_dir+'XC'))-0.1
    ZC = np.squeeze(rdmds(_dir+'RC'))+0.4
    T0=np.squeeze(rdmds(_dir+'T',0))
    
    def init():
        cs1 = ax.contour(XC,ZC,T0)

    def update(n): 
        # n = frame counter
        T=np.squeeze(rdmds(_dir+'T',n*doutcase))
        cs1 = ax.contour(XC,ZC,T)

        ax.cla()
        ax.contourf(XC,ZC,T-T0,150,vmin=-0.002,vmax=.002)
        ax.set_xlim([-0.1, 0.3])
        ax.set_ylim([0, 0.4])
        lvls = np.linspace(10,10.02,20)
        ax.contour(XC,ZC,T,lvls,colors=['k'])

    anim = animation.FuncAnimation(fig, update, init_func=init, frames=int(maxoutcase/doutcase), blit=True)

    # anim.save can be called in a few different ways, some which might or might not work
    # on different platforms and with different versions of matplotlib and video encoders
    anim.save(fname, fps=5, writer="ffmpeg", codec="libx264")

    plt.close(fig)
```

```python
%%capture --no-stdout
gen_anim("/ocean/rirwin/Half_Cylinder_MITgcm/Stratified_Cylinder/N0p01/STR01280128/",1000,600000,"N0p01_STR01280128.mp4")
print "Done 1"
gen_anim("/ocean/rirwin/Half_Cylinder_MITgcm/Stratified_Cylinder/N0p01/STR02560256/",1000,600000,"N0p01_STR02560256.mp4")
print "Done 2"
gen_anim("/ocean/rirwin/Half_Cylinder_MITgcm/Stratified_Cylinder/N0p01/STR05120512/",1000,600000,"N0p01_STR05120512.mp4")
print "Done 3"
gen_anim("/ocean/rirwin/Half_Cylinder_MITgcm/Stratified_Cylinder/N0p10/STR01280128/",1000,600000,"N0p10_STR01280128.mp4")
print "Done 4"
gen_anim("/ocean/rirwin/Half_Cylinder_MITgcm/Stratified_Cylinder/N0p10/STR02560256/",1000,600000,"N0p10_STR02560256.mp4")
print "Done 5"
```
