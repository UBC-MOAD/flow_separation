The state of the flow-over cylinder problem in the MITgcm.

```python
from __future__ import division # fix division as floating point division
import numpy as np # get numpy
import matplotlib.pyplot as plt # get plotting
from MITgcmutils import rdmds
import matplotlib.pylab as pylab

# plot inline
%matplotlib inline 
```

The goal of this analysis is to quantify how well the MITgcm handles flow-over topography. This is done by comparing a well established theoretical solution to numerical output.

Initially the theoretical background for this problem was work done by Blausius and others in the first half of the 20th century --- this work focussed on using the potential flow solution to derive the surface pressure term used in Prandtl's Boundary Layer equations. This work produced theoretical predictions for position of boundary layer separation, and it was this value that was to be used in the current problem.

There are issues with comparing the theoretical approach to the numerical output:

1) the potential flow problem is solved on an infinite domain

  - the numerical code must be solved on a finite domain with specified BCs
  
2) the potential flow solution is steady and inviscid

  - the steady state solution from potential flow theory is not consistent with a tank-scale setup in the MITgcm
  - even if the code is seeded with the potential flow solution, the viscous and unsteady nonlinear dynamics quickly adjust the field

```python
X = np.squeeze(rdmds('../../Cyl_Re/Nx512Nz256_Re500_Seed/XC'))-0.3
Z = np.squeeze(rdmds('../../Cyl_Re/Nx512Nz256_Re500_Seed/RC'))+0.4
U00 = np.squeeze(rdmds('../../Cyl_Re/Nx512Nz256_Re500_Seed/U',0))
U01 = np.squeeze(rdmds('../../Cyl_Re/Nx512Nz256_Re500_Seed/U',200))
U02 = np.squeeze(rdmds('../../Cyl_Re/Nx512Nz256_Re500_Seed/U',400))

pylab.rcParams['figure.figsize'] = (16,4)

plt.figure()
plt.subplot(1,3,1)
plt.pcolor(X,Z,U00,cmap='Spectral_r')
plt.xlim((-0.05,0.05))
plt.ylim((0,0.1))
plt.title('Potential solution')

plt.subplot(1,3,2)
plt.pcolor(X,Z,U01,cmap='Spectral_r')
plt.xlim((-0.05,0.05))
plt.ylim((0,0.1))
plt.title('MITgcm output, t=1s')

plt.subplot(1,3,3)
plt.pcolor(X,Z,U02,cmap='Spectral_r')
plt.xlim((-0.05,0.05))
plt.ylim((0,0.1))
plt.title('MITgcm output, t=2s')
plt.show()
```

As can be seen from the figure above, with a Reynolds number of $Re=500$ the potential flow solution is immediately modified by the numerics.

In order to look at a true potential flow field, the code would have to be initialized with a much higher Reynolds number in a much larger domain. 

It should also be noted that the MITgcm did a poor job representing the $w$ field (vertical velocity). This vertical velocity is calculated from the $u$ initial field, and so should have produced the potential flow field, however there were some errors:

```python
W00 = np.squeeze(rdmds('../../Cyl_Re/Nx512Nz256_Re500_Seed/W',0))

pylab.rcParams['figure.figsize'] = (6,5)

plt.figure()
plt.pcolor(X,Z,W00,cmap='Spectral_r')
plt.xlim((-0.05,0.05))
plt.ylim((0,0.1))
plt.title('MITgcm calculated $w$ field')
plt.colorbar()
plt.show()
```

These vertical bands are clearly sources of error in the initialization. This can be remedied by prescribing the $w$ field in a non-hydrostatic run. However, doing this seems fruitless -- the impulsively started run does not adjust to the potential field. Thus, it seems unlikely that a steady-state solution in a viscous numerical run would approach the potential field.

Instead of going this route, a new means of quantifying the MITgcm's handling of boundaries is proposed:

- 1) impulsively start an "inviscid" tank-scale run (note that inviscid means using free-slip conditions, as truly invscid numerical runs are unstable)

- 2) wait until the output adjusts to a steady state

- 3) using this steady state, a pressure field at the surface can be inferred from the data using Bernoulli's principle

- 4) this pressure field at the surface can then be used in tandem with the theoretical Prandtl's Boundary Layer equations to approximate the first N terms of the series expansion solution (as was done by Blausius for symmetric and asymmetric pressure fields)

- 5) using this solution, a separation point may be determined (along with structure of the boundary layer to an extent)

- 6) a viscous run (no-slip conditions) seeded with the inviscid steady-state will show the numerical evolution of the boundary layer and the results can be compared with the theoretical solution

Number 4 will use Matlab or Python code to numerically evaluate the N nonlinear BVP's (using methods similar to Blausius and others). Formulating these equations and writing this code will be the most intensive aspect of this project.
