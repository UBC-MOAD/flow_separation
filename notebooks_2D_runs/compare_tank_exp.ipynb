```python
ncdata_cstate = nc.Dataset('../../../DAWE_COMP/Nx256Ny268Nz182_spinup/glue_state.nc')
ncdata_cgrid  = nc.Dataset('../../../DAWE_COMP/Nx256Ny268Nz182_spinup/glue_grid.nc')

ncvar_cThetaC = ncdata_cgrid.variables['X']
ncvar_crC = ncdata_cgrid.variables['Y']
ncvar_cZC = ncdata_cgrid.variables['Z']
ncvar_cDepth = ncdata_cgrid.variables['Depth']

cDepth = np.array(ncvar_cDepth)

cThetaC = np.array(ncvar_cThetaC)/(360.0)*2*np.pi
crC = np.array(ncvar_crC)
cZC = np.array(ncvar_cZC)

[cm_ThetaC,cm_rC]=np.meshgrid(cThetaC,crC)

ncvar_cUtheta = ncdata_cstate.variables['U']
ncvar_cUr = ncdata_cstate.variables['V']
ncvar_cW = ncdata_cstate.variables['W']
ncvar_cS = ncdata_cstate.variables['S']

ncdata_ogrid = nc.Dataset('../../../DAWE_COMP/hires_merge/grid.nc')
ncvar_oThetaC = ncdata_ogrid.variables['X']
ncvar_orC = ncdata_ogrid.variables['Y']
ncvar_oZC = ncdata_ogrid.variables['Z']
ncvar_oDepth = ncdata_ogrid.variables['Depth']

oDepth = np.array(ncvar_oDepth)

oThetaC = np.array(ncvar_oThetaC)/(360.0)*2*np.pi
orC = np.array(ncvar_orC)
oZC = np.array(ncvar_oZC)

[om_ThetaC,om_rC]=np.meshgrid(oThetaC,orC)

ncdata_oUtheta = nc.Dataset('../../../DAWE_COMP/hires_merge/U.nc')
ncvar_oUtheta = ncdata_oUtheta.variables['U']
ncdata_oUr = nc.Dataset('../../../DAWE_COMP/hires_merge/V.nc')
ncvar_oUr = ncdata_oUr.variables['V']
ncdata_oW = nc.Dataset('../../../DAWE_COMP/hires_merge/W.nc')
ncvar_oW = ncdata_oW.variables['W']
ncdata_oS = nc.Dataset('../../../DAWE_COMP/hires_merge/S.nc')
ncvar_oS = ncdata_oS.variables['S']
```

Convert cylindrical coordinate meshes into cartesian coordinate meshes for plotting

```python
cm_XC = cm_rC*np.cos(cm_ThetaC)
cm_YC = cm_rC*np.sin(cm_ThetaC)

om_XC = om_rC*np.cos(om_ThetaC)
om_YC = om_rC*np.sin(om_ThetaC)
```

```python
def convert_polar_velocity_to_cartesian(U_r,U_theta,rC,thetaC):
    U = U_r*np.cos(thetaC)-rC*U_theta*np.sin(thetaC)
    V = U_r*np.sin(thetaC)+rC*U_theta*np.cos(thetaC)
    
    return [U,V]
```

```python
oOut = 1
cOut = 1
lvl = 10

cUtheta = np.array(ncvar_cUtheta[cOut,lvl,:,1:])
cUr = np.array(ncvar_cUr[cOut,lvl,1:,:])

[cU,cV] = convert_polar_velocity_to_cartesian(cUr,cUtheta,cm_rC,cm_ThetaC)

oUtheta = np.array(ncvar_oUtheta[oOut,lvl,:,1:])
oUr = np.array(ncvar_oUr[oOut,lvl,1:,:])

[oU,oV] = convert_polar_velocity_to_cartesian(oUr,oUtheta,om_rC,om_ThetaC)
```

```python
dx = 8
dy = 8

plt.subplot(2,1,1)
plt.quiver(cm_XC[dy::dy,dx::dx],cm_YC[dy::dy,dx::dx],np.squeeze(cU[dy::dy,dx::dx]),np.squeeze(cV[dy::dy,dx::dx]))
plt.contour(cm_XC,cm_YC,cDepth,colors='k')
plt.xlim((-0.4,-0.2))
plt.ylim((-0.1,0.1))
plt.title('Current MITgcm output')

plt.subplot(2,1,2)
plt.quiver(om_XC[dy::dy,dx::dx],om_YC[dy::dy,dx::dx],np.squeeze(oU[dy::dy,dx::dx]),np.squeeze(oV[dy::dy,dx::dx]))
plt.contour(om_XC,om_YC,oDepth,colors='k')
plt.xlim((-0.4,-0.2))
plt.ylim((-0.1,0.1))
plt.title('Old MITgcm output')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.pcolor(om_XC,om_YC,oU)
plt.subplot(1,2,2)
plt.pcolor(om_XC,om_YC,oV)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.pcolor(cm_XC,cm_YC,cU)
plt.subplot(1,2,2)
plt.pcolor(cm_XC,cm_YC,cV)
plt.show()
```
