This notebook compares flow-over cylinder runs at different resolutions. Compares separation point as a function of time.

```python
import cylinder_utils as cu
import numpy as np
import matplotlib.pyplot as plt
from MITgcmutils import rdmds

import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = (16,12)

from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

%matplotlib inline
```

The following cell loads in the grid coordinates and sets the variable run parameters.

```python
Ncomps = 3
Nx = [256,1024,1024]
Nz = [64,128,256]

dirs = ['../../CylinderNx256Nz64_001','../../CylinderNx1024Nz128_002','../../CylinderNx1024Nz256_001']

X=[0]*Ncomps
Z=[0]*Ncomps
dt=[0]*Ncomps
R0=[0]*Ncomps
Lx=[0]*Ncomps
Lz=[0]*Ncomps

dt[0] = 4e-3
dt[1] = 1e-3
dt[2] = 2.5e-4

Lx[0] = 0.4
Lx[1] = 0.4
Lx[2] = 0.4

Lz[0] = 0.1
Lz[1] = 0.1
Lz[2] = 0.1

R0[0] = 0.025
R0[1] = 0.025
R0[2] = 0.025

del_out = [0]*Ncomps
max_out = [0]*Ncomps

tmax = 35.0

for ii in range(0,Ncomps):
    del_out[ii] = int(1.0/dt[ii])
    max_out[ii] = int(tmax/dt[ii])
    X[ii] = np.squeeze(rdmds(dirs[ii]+'/XC')-Lx[ii]/2)
    Z[ii] = np.squeeze(rdmds(dirs[ii]+'/RC')+Lz[ii])
```

This cell confirms that the calc_dvtheta_dr function is working correctly; debug cell

```python
jj = 0
T=np.squeeze(rdmds(dirs[jj]+'/T',0))
bd_pts = cu.find_bdy(X[jj],Z[jj],R0[jj],T)
tmplst = []
U=np.squeeze(rdmds(dirs[jj]+'/U',250))
W=np.squeeze(rdmds(dirs[jj]+'/W',250))
[theta_bd,dvtheta_dr_bd] = cu.calc_dvtheta_dr(X[jj],Z[jj],U,W,R0[jj],Nx[jj],Nz[jj],bd_pts)
plt.plot(theta_bd,dvtheta_dr_bd)
plt.plot(theta_bd,np.zeros(len(theta_bd)),'--k')
plt.ylim((-200,200))
plt.show()
```

This cell does the bulk of the work. Calculates the separation point at each output for each comparison directory

```python
sep_pt = [0]*Ncomps
for jj in range(0,Ncomps):
    T=np.squeeze(rdmds(dirs[jj]+'/T',0))
    bd_pts = cu.find_bdy(X[jj],Z[jj],R0[jj],T)
    tmplst = []
    for ii in range(0,max_out[jj]+1,del_out[jj]):
        U=np.squeeze(rdmds(dirs[jj]+'/U',ii))
        W=np.squeeze(rdmds(dirs[jj]+'/W',ii))
        [theta_bd,dvtheta_dr_bd] = cu.calc_dvtheta_dr(X[jj],Z[jj],U,W,R0[jj],Nx[jj],Nz[jj],bd_pts)
        tmplst.append(cu.find_zero(theta_bd,dvtheta_dr_bd))
    sep_pt[jj] = tmplst
```

This cell simply plots the data calculated in the previous cell

```python
form = ['-or','-ob','-og']
for ii in range(Ncomps):
    plt.plot(sep_pt[ii],form[ii])
plt.legend(('256x64','1024x128','1024x256'))
plt.xlabel('time in seconds')
plt.ylabel('separation angle')
plt.show()
```

This cell calculates the flow potential, phi.

```python
def calc_phi(X,Z,U,W,bd_pts):
    phix = np.zeros((len(Z),len(X)))
    for zind, zz in enumerate(Z):
        sumsofar = 0
        for xind, xx in enumerate(X):
            sumsofar = sumsofar + U[zind,xind]
            phix[zind,xind] = sumsofar
            
    phix = phix*(X[1]-X[0])
    
    phiz = np.zeros((len(Z),len(X)))
    for xind, xx in enumerate(X):
        sumsofar = 0
        for zind, zz in enumerate(Z):
            sumsofar = sumsofar + W[zind,xind]
            phiz[zind,xind] = sumsofar
    
    phiz = -phiz*(Z[1]-Z[0])
    
    phi = phix + phiz
    
    phi = phi - np.mean(phi)
    
    return phi
```

```python
sel = 1

U=np.squeeze(rdmds(dirs[sel]+'/U',1*del_out[sel]))
W=np.squeeze(rdmds(dirs[sel]+'/W',1*del_out[sel]))
T = np.squeeze(rdmds(dirs[sel]+'/T',0))
bd_pts = cu.find_bdy(X[sel],Z[sel],R0[sel],T)

phi = calc_phi(X[sel],Z[sel],U,W,bd_pts)

plt.pcolor(X[sel],Z[sel],phi)
plt.colorbar()
plt.show()
```

```python
def calc_dx2(U):
    uext_r = U[:,1:]
    uext_r = np.append(uext_r, 0.01+np.zeros((256,1)), 1)
    uext_l = U[:,0:-1]
    uext_l = np.append(0.01+np.zeros((256,1)),uext_l, 1)
    
    print uext_r.shape
    print uext_l.shape
    print U.shape
    
    u_xx = (uext_r-2*U+uext_l)
    return u_xx
    
```

```python
def calc_dz2(U):
    uext_u = U[1:,:]
    uext_u = np.append(uext_u, np.zeros((1,1024)), 0)
    uext_d = U[0:-1,:]
    uext_d = np.append(np.zeros((1,1024)),uext_d, 0)
    
    print uext_u.shape
    print uext_d.shape
    print U.shape
    
    u_zz = (uext_u-2*U+uext_d)
    return u_zz
```

```python
sel = 2

U=np.squeeze(rdmds(dirs[sel]+'/U',15*del_out[sel]))
W=np.squeeze(rdmds(dirs[sel]+'/W',15*del_out[sel]))
T = np.squeeze(rdmds(dirs[sel]+'/T',0))
bd_pts = cu.find_bdy(X[sel],Z[sel],R0[sel],T)

phi = calc_phi(X[sel],Z[sel],U,W,bd_pts)

phi_xx = calc_dx2(phi)
phi_zz = calc_dz2(phi)
```
