This notebook compares several 2D flow separation runs.

```python
from __future__ import division # fix division as floating point division
import numpy as np
import matplotlib.pyplot as plt
import process_data as pd 
from MITgcmutils import rdmds

# plot inline
%matplotlib inline 

# change default size of figures
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = (16,12)

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

Load in two data sets from two directories.

```python
_Ncomps = 4; # number of comparisons

# setup all of the default tuple objects
_Nx = [0]*_Ncomps; _Nz = [0]*_Ncomps;
_total_depth = [0]*_Ncomps;
_dir = ['']*_Ncomps;
_title = ['']*_Ncomps;
_dt = [0]*_Ncomps;
_x = [0]*_Ncomps;
_z = [0]*_Ncomps;
_depth = [0]*_Ncomps;

_U = [0]*_Ncomps;
_V = [0]*_Ncomps;
_W = [0]*_Ncomps;
_T = [0]*_Ncomps;
```

Now that we have the correct structures, input each parameter set.

```python
# now set each run specific parameter set
_Nx[0] = 256; _Nx[1] = 256; _Nx[2] = 384; _Nx[3] = 512;
_Nz[0] = 64; _Nz[1] = 256; _Nz[2] = 256; _Nz[3] = 256;

# total depth in metres
_total_depth[0] = 200; _total_depth[1] = 200; _total_depth[2] = 200; _total_depth[3] = 200;

# directories
_dir[0] = '../../Orl_Ext_Nx_%d_Nz_%d' % (_Nx[0],_Nz[0]);
_dir[1] = '../../Orl_Ext_Nx_%d_Nz_%d' % (_Nx[1],_Nz[1]);
_dir[2] = '../../Orl_Ext_Nx_%d_Nz_%d' % (_Nx[2],_Nz[2]);
_dir[3] = '../../MPIOrl_Ext_Nx%dNz%d' % (_Nx[3],_Nz[3]);
#_dir[1] = '../../FreeSlip_Nx_%d_Nz_%d' % (_Nx[1],_Nz[1]);
#_dir[2] = '../../QDrag0p002_Nx_%d_Nz_%d' % (_Nx[2],_Nz[2]);
#_dir[3] = '../../QDrag0p100_Nx_%d_Nz_%d' % (_Nx[3],_Nz[3]);

# used for titling plots
_title[0] = r'$Nx:$ %d, $Nz:$ %d' % (_Nx[0],_Nz[0])
_title[1] = r'$Nx:$ %d, $Nz:$ %d' % (_Nx[1],_Nz[1])
_title[2] = r'$Nx:$ %d, $Nz:$ %d' % (_Nx[2],_Nz[2])
_title[3] = r'$Nx:$ %d, $Nz:$ %d' % (_Nx[3],_Nz[3])


# loop through and read/generate the x and z coordinate arrays
for _ii in range(0,_Ncomps):
    _x[_ii] = rdmds(_dir[_ii]+'/XG');
    _z[_ii] = np.linspace(-_total_depth[_ii],0,_Nz[_ii]);
    _z[_ii] = _z[_ii][::-1];
    _depth[_ii] = np.squeeze(rdmds(_dir[_ii]+'/Depth'));
    
# timestepping parameters RLI: load from data file? probably unnecessary
_dt[0] = 10; _dt[1] = 1; _dt[2] = 1; _dt[3] = 1;
_output_dt = 1800;
_final_t = 75600+_output_dt;
```

Now we have all of the data necessary to load the files in to iPython. Use the process_data script function load_UVWT to accomplish this.

```python
# now load in all of the necessary files
_out_files = [0]*_Ncomps;
for _ii in range(0,_Ncomps):
    _out_files[_ii] = np.arange(0,_final_t/_dt[_ii],_output_dt/_dt[_ii]);
    [_U[_ii],_V[_ii],_W[_ii],_T[_ii]] = pd.load_UVWT(_Nx[_ii],_Nz[_ii],_dir[_ii],_out_files[_ii]);
```

Now that we have the files read in, we can compare some fields.

```python
_fig = plt.figure();

plt.clf();

_outs = [10,20,30,len(_out_files[0])-1]; # set the outputs to compare --- can be variable length

# plot each subplot
_itr = 1;
for _ii in range(0,_Ncomps):
    for _jj,_out in enumerate(_outs):
        plt.subplot(_Ncomps,len(_outs),_itr);
        plt.pcolor(_x[_ii]/1000,_z[_ii],np.squeeze(_U[_ii][:,:,_out]),cmap='Spectral_r');
        plt.colorbar();
        plt.clim((-0.05, 0.3));
        plt.plot(np.squeeze(_x[_ii])/1000,-_depth[_ii],'-k')
        plt.xlabel(r'$x$ in km')
        plt.ylabel(r'$z$ in m')
        plt.xlim((0,15))
        plt.ylim((-_total_depth[_ii],0))
        plt.title(_title[_ii]+', output: '+str(_out));
        _itr = _itr+1;

plt.suptitle(r'Comparing $u$ fields at various outputs', fontsize=14, fontweight='bold')
        
_fig.subplots_adjust(hspace=0.4,wspace=0.2);

plt.show();
```

Compare the grid spacing around the slope in each simulation.

```python
_fig = plt.figure();

for _ii in range(0,_Ncomps):
    plt.subplot(_Ncomps,1,_ii+1);
    plt.plot(np.squeeze(_x[_ii]),-_depth[_ii],'-b',linewidth=2,drawstyle='steps');
    plt.plot(np.squeeze(_x[_ii]),np.tile(_z[_ii],[_Nx[_ii],1]),'-k');
    plt.plot(np.tile(np.squeeze(_x[_ii]),[_Nz[_ii],1]),_z[_ii],'-k');
    plt.title(_title[_ii]+' grid spacing');
    plt.xlabel('x'); plt.ylabel('z');
    plt.xlim((4e3,7e3));
    plt.ylim((-65,-45));
    
_fig.subplots_adjust(hspace=0.4,wspace=0.2);
```
