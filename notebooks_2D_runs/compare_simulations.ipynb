This notebook compares two 2D flow separation runs.

```python
from __future__ import division # fix division as floating point division
import numpy as np
import matplotlib.pyplot as plt
import process_data as pd 
from MITgcmutils import rdmds

# plot inline
%matplotlib inline 

# change default size of figures
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = (16,12)
```

Load in two data sets from two directories.

```python
_Nx1 = 256; _Nx2 = 256;
_Ny = 1;
_Nz1 = 64; _Nz2 = 64;

_total_depth1 = 200;
_total_depth2 = 200;

_dir1 = '../../Orlanski_Nx_%d_Nz_%d' % (_Nx1,_Nz1);
_dir2 = '../../Nx_%d_Nz_%d' % (_Nx2,_Nz2);

_dt1 = 10; _dt2 = 10;
_output_dt = 1800;
_final_t = 86400+_output_dt;

_out_files1 = np.arange(0,_final_t/_dt1,_output_dt/_dt1);
_out_files2 = np.arange(0,_final_t/_dt2,_output_dt/_dt2);

_z1 = np.linspace(-_total_depth1,0,_Nz1);
_z2 = np.linspace(-_total_depth1,0,_Nz2);
_z1 = _z1[::-1];
_z2 = _z2[::-1];

_x1 = np.squeeze(rdmds(_dir1+'/XC'));
_x2 = np.squeeze(rdmds(_dir2+'/XC'));
_depth1 = np.squeeze(rdmds(_dir1+'/Depth'));
_depth2 = np.squeeze(rdmds(_dir2+'/Depth'));

[_U1,_V1,_W1,_T1] = pd.load_UVWT(_Nx1,_Nz1,_dir1,_out_files1);
[_U2,_V2,_W2,_T2] = pd.load_UVWT(_Nx2,_Nz2,_dir2,_out_files2);
```

Now that we have the files read in, we can compare some fields.

```python
_fig = plt.figure();

plt.subplot(2,2,1);
plt.pcolor(_x1/1000,_z1,np.squeeze(_U1[:,:,1]),cmap='Spectral_r');
plt.colorbar();
plt.contour(_x1/1000,_z1,np.squeeze(_U1[:,:,1]),10,cmap='binary_r');
plt.title('U Orlanski east BC, 0.5 hours')

plt.subplot(2,2,2);
plt.pcolor(_x2/1000,_z2,np.squeeze(_U2[:,:,1]),cmap='Spectral_r');
plt.colorbar();
plt.contour(_x2/1000,_z2,np.squeeze(_U2[:,:,1]),10,cmap='binary_r');
plt.title('U default east BC, 0.5 hours')

plt.subplot(2,2,3);
plt.pcolor(_x1/1000,_z1,np.squeeze(_U1[:,:,47]),cmap='Spectral_r');
plt.colorbar();
plt.contour(_x1/1000,_z1,np.squeeze(_U1[:,:,47]),10,cmap='binary_r');
plt.title('U Orlanski east BC, 24 hours')

plt.subplot(2,2,4);
plt.pcolor(_x2/1000,_z2,np.squeeze(_U2[:,:,47]),cmap='Spectral_r');
plt.colorbar();
plt.contour(_x2/1000,_z2,np.squeeze(_U2[:,:,47]),10,cmap='binary_r');
plt.title('U default east BC, 24 hours')

plt.show();
```

```python
_fig = plt.figure();

_T1plt = _T1; _T2plt = _T2;
_T1plt[_T1plt<10] = 10;
_T2plt[_T2plt<10] = 10;

plt.subplot(2,2,1);
plt.pcolor(_x1/1000,_z1,np.squeeze(_T1[:,:,1]),cmap='Spectral_r');
plt.colorbar(); plt.clim((10,14))
plt.contour(_x1/1000,_z1,np.squeeze(_T1plt[:,:,1]),10,cmap='binary_r');
plt.title('T Orlanski east BC, 0.5 hours')

plt.subplot(2,2,2);
plt.pcolor(_x2/1000,_z2,np.squeeze(_T2[:,:,1]),cmap='Spectral_r');
plt.colorbar(); plt.clim((10,14))
plt.contour(_x2/1000,_z2,np.squeeze(_T2plt[:,:,1]),10,cmap='binary_r');
plt.title('T default east BC, 0.5 hours')

plt.subplot(2,2,3);
plt.pcolor(_x1/1000,_z1,np.squeeze(_T1[:,:,47]),cmap='Spectral_r');
plt.colorbar(); plt.clim((10,14))
plt.contour(_x1/1000,_z1,np.squeeze(_T1plt[:,:,47]),10,cmap='binary_r');
plt.title('T Orlanski east BC, 24 hours')

plt.subplot(2,2,4);
plt.pcolor(_x2/1000,_z2,np.squeeze(_T2[:,:,47]),cmap='Spectral_r');
plt.colorbar(); plt.clim((10,14))
plt.contour(_x2/1000,_z2,np.squeeze(_T2plt[:,:,47]),10,cmap='binary_r');
plt.title('T default east BC, 24 hours')

plt.show();
```
