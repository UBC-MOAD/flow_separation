This code compares the results of the free-slip and no-slip-over-cylinder-runs.

```python
import numpy as np
import matplotlib.pyplot as plt
from MITgcmutils import rdmds

import matplotlib.pylab as pylab

%matplotlib inline
```

```python
x = np.squeeze(rdmds('../Free_Slip_Cylinder/run1/XC'))
z = np.squeeze(rdmds('../Free_Slip_Cylinder/run1/RC'))
```

```python
UFS = np.squeeze(rdmds('../Free_Slip_Cylinder/run1/U',2000))
UNS = np.squeeze(rdmds('../No_Slip_Cylinder/run1/U',2000))

pylab.rcParams['figure.figsize'] = (20,8)

plt.figure()
plt.subplot(1,3,1)
plt.pcolor(x,z,UFS,cmap='Spectral_r')
plt.colorbar()
plt.xlim((0.15,0.25))
plt.ylim((-0.2,-0.15))
plt.title('Free slip U output')

plt.subplot(1,3,2)
plt.pcolor(x,z,UNS,cmap='Spectral_r')
plt.colorbar()
plt.xlim((0.15,0.25))
plt.ylim((-0.2,-0.15))
plt.title('No slip U output')

plt.subplot(1,3,3)
plt.pcolor(x,z,UNS-UFS,cmap='Spectral_r')
plt.colorbar()
plt.xlim((0.15,0.25))
plt.ylim((-0.2,-0.15))
plt.title('No slip U minus free slip U')

plt.suptitle('Early time output comparison')
plt.show()
```

```python
UFS = np.squeeze(rdmds('../Free_Slip_Cylinder/run1/U',16400))
UNS = np.squeeze(rdmds('../No_Slip_Cylinder/run1/U',16400))

pylab.rcParams['figure.figsize'] = (20,8)

plt.figure()
plt.subplot(1,3,1)
plt.pcolor(x,z,UFS,cmap='Spectral_r')
plt.colorbar()
plt.xlim((0.15,0.25))
plt.ylim((-0.2,-0.15))
plt.title('Free slip U output')

plt.subplot(1,3,2)
plt.pcolor(x,z,UNS,cmap='Spectral_r')
plt.colorbar()
plt.xlim((0.15,0.25))
plt.ylim((-0.2,-0.15))
plt.title('No slip U output')

plt.subplot(1,3,3)
plt.pcolor(x,z,UNS-UFS,cmap='Spectral_r')
plt.colorbar()
plt.xlim((0.15,0.25))
plt.ylim((-0.2,-0.15))
plt.title('No slip U minus free slip U')

plt.suptitle('Later time output comparison')
plt.show()
```
