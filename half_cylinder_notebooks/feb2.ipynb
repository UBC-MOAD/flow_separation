```python
%%capture --no-stdout
cyl_case = []

#cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Unstratified_Cylinder/Resolution_Dependence/Re25_Final/Nx064Nz064/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Unstratified_Cylinder/Resolution_Dependence/Re25_Final/Nx128Nz128/'))
#cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Unstratified_Cylinder/Resolution_Dependence/Re25_Final/Nx128Nz128_SEED/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Unstratified_Cylinder/Resolution_Dependence/Re25_Final/Nx256Nz256/'))
#cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Unstratified_Cylinder/Resolution_Dependence/Re25_Final/Nx128Nz128_IRR/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Unstratified_Cylinder/Resolution_Dependence/Re25_Final/Nx512Nz512/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Unstratified_Cylinder/Resolution_Dependence/Re25_Final/Nx256Nz256_IRR/'))
#cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Unstratified_Cylinder/Resolution_Dependence/Re25_Final/Nx512Nz512_IRR/'))
#cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re25_Impulse/Nx512Nz512_IRR2/'))
```

```python
NCASES = len(cyl_case)

dout = []
max_out = []
id_str = []
dx_vals = []
dz_vals = []

#Re = 25

#delta = np.sqrt(0.025*2e-5/(2*0.01))
delta = 0.025

for ii in range(0,NCASES):
    outdat = cyl_case[ii].find_out_data()
    dout.append(outdat[0])
    max_out.append(outdat[1])
    dx_vals.append(np.min(np.abs(cyl_case[ii].XC[1:]-cyl_case[ii].XC[:-1])))
    dz_vals.append(np.min(np.abs(cyl_case[ii].ZC[1:]-cyl_case[ii].ZC[:-1])))
    id_str.append(r"$\delta$/dz={:1.0f}".format(np.floor(delta/dz_vals[ii])))

print id_str
```

```python
%%capture --no-stdout
all_sep_pts = []
for jj in range(0,NCASES):
    sep_pts = []
    for ii in range(0,int(max_out[jj]/dout[jj])+1):
        if ii*dout[jj] <= max_out[jj]:
            thet_tmp,dudz_tmp = cyl_case[jj].get_dudz(dout[jj]*ii)
            sep_pts.append([ii*dout[jj],cyl_case[jj].find_zero(thet_tmp,dudz_tmp)])
    all_sep_pts.append(sep_pts)
    print "Done case " + str(jj) + "."

```

```python
pylab.rcParams['figure.figsize'] = (9.0, 6.0)

plt.figure()

cols = [[.3,.3,.9],[.9,.3,.3],[.3,.9,.3],[.6,.6,.3]]

fin_sep = []
for ii in range(0,NCASES):
    sep_pts = all_sep_pts[ii]
    tt = []
    sep = []
    for jj in sep_pts:
        tt.append((jj[0]/dout[ii])/(np.pi*0.025/0.01))
        sep.append(jj[1])
    fin_sep.append(sep[-1])
    plt.plot(sep,tt,'-',label=id_str[ii],linewidth=2)
    
plt.xlim((120,180))
plt.ylim((0,10))
#plt.plot([100,180],[0.35,0.35],'-k',label=r'$t_s$')
#plt.plot([104.5,104.5],[0,20],'--k',label=r'Blasius sep.')
#plt.fill_betweenx([0,10],122,135,facecolor=[1,1,1],label='Wu region')
#plt.plot([124,124],[0,20],':',label=r'Wu sep. Re=50',linewidth=3,color='k')
plt.plot([132.53999999999999,132.53999999999999],[0,20],'-.',label=r'Wu sep. Re=25',linewidth=3,color='k')
plt.fill_betweenx([0,10],130,142,facecolor=[1,1,1],label='Wu region, Re=25')
#plt.plot([135,135],[0,20],'--g',label=r'Wu sep. Re=50')
plt.legend(loc=1,fontsize=12)
plt.ylabel(r'$t^*$',fontsize=24)
plt.xlabel(r'$\theta$',fontsize=24)
plt.show()
```

```python
def gen_anim(cylcase,doutcase,maxoutcase,fname):
    fig, ax = plt.subplots(figsize=(10,10))

    ax.set_xlim([-0.1, 0.3])
    ax.set_ylim([0, 0.4])

    def init():
        U,W,vort = cylcase.get_fields_at_output(0)
        cs1 = ax.contour(cylcase.XC,cylcase.ZC,vort)

    def update(n): 
        # n = frame counter
        U,W,vort=cylcase.get_fields_at_output(n*doutcase)
        thet,dudz = cylcase.get_dudz(n*doutcase); 
        sep_pt = cylcase.find_zero(thet,dudz)

        thet_0 = (180-sep_pt)*np.pi/180

        ax.cla()
        ax.pcolor(cylcase.XC,cylcase.ZC,vort,cmap='Spectral_r')
        ax.fill(cylcase.XG,0.4-cylcase.Depth,facecolor='gray',lw=3)
        ax.plot([0, np.cos(thet_0)*0.025],[0,np.sin(thet_0)*0.025],'-w')
        ax.set_xlim([-0.1, 0.3])
        ax.set_ylim([0, 0.4])
        ax.contour(cylcase.XC,cylcase.ZC,vort,levels=[0],colors=['k'])

    anim = animation.FuncAnimation(fig, update, init_func=init, frames=int(maxoutcase/doutcase), blit=True)

    # anim.save can be called in a few different ways, some which might or might not work
    # on different platforms and with different versions of matplotlib and video encoders
    anim.save(fname, fps=5, writer="ffmpeg", codec="libx264")

    plt.close(fig)
```

```python
%%capture --no-stdout
for ii in range(0,NCASES):
    gen_anim(cyl_case[ii],dout[ii],min(40*dout[ii],max_out[ii]),str(len(cyl_case[ii].XC))+str(len(cyl_case[ii].ZC))+".mp4")
    print "Done case " + str(ii+1)
```
