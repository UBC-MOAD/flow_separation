```python
NCASES = 5
dout = []
max_out = []

id_str = []

#Re = 25
dout.append(1000)
dout.append(2000)
dout.append(5000)
dout.append(5000)
dout.append(10000)
max_out.append(80000)
max_out.append(160000)
max_out.append(400000)
max_out.append(300000)
max_out.append(330000)

delta = np.sqrt(2e-5/0.01)

id_str.append("dx={:1.3f}".format(0.4/128/delta)+'$\delta$, '+"dz={:1.3f}".format(0.4/128/delta)+'$\delta$')
id_str.append("dx={:1.3f}".format(0.4/256/delta)+'$\delta$, '+"dz={:1.3f}".format(0.4/128/delta)+'$\delta$')
id_str.append("dx={:1.3f}".format(0.4/256/delta)+'$\delta$, '+"dz={:1.3f}".format(0.4/256/delta)+'$\delta$')
id_str.append("dx={:1.3f}".format(0.4/512/delta)+'$\delta$, '+"dz={:1.3f}".format(0.4/256/delta)+'$\delta$')
id_str.append("dx={:1.3f}".format(0.4/512/delta)+'$\delta$, '+"dz={:1.3f}".format(0.4/512/delta)+'$\delta$')
#id_str.append('Re25.128x128')
#id_str.append('Re25.256x128')
#id_str.append('Re25.256x256')
#id_str.append('Re25.512x256')
#id_str.append('Re25.512x512')
```

```python
%%capture --no-stdout
all_sep_pts = []
for jj in range(0,NCASES):
    sep_pts = []
    for ii in range(0,int(max_out[jj]/dout[jj])+1):
        if ii*dout[jj] <= max_out[jj]:
            thet_tmp,dudz_tmp = cyl_case[jj].get_dudz(dout[jj]*ii)
            sep_pts.append([ii*dout[jj],cyl_case[jj].find_zero(thet_tmp,dudz_tmp)])
    all_sep_pts.append(sep_pts)
    print "Done case " + str(jj) + "."
```

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

plt.figure()

for ii in range(0,NCASES):
    sep_pts = all_sep_pts[ii]
    tt = []
    sep = []
    for jj in sep_pts:
        tt.append((jj[0]/dout[ii])/(np.pi*0.025/0.01))
        sep.append(jj[1])
    plt.plot(sep,tt,'-o',label=id_str[ii])
    
plt.xlim((120,160))
plt.ylim((0,10))
plt.plot([100,180],[0.35,0.35],'-k',label=r'$t_s$')
#plt.plot([104.5,104.5],[0,20],'--k',label=r'Blasius sep.')
plt.fill_betweenx([0,10],122,135,facecolor='gray',label='Wu region')
plt.plot([124,124],[0,20],'--b',label=r'Wu sep. Re=50')
#plt.plot([135,135],[0,20],'--g',label=r'Wu sep. Re=50')
plt.legend(loc=2)
plt.grid()
plt.ylabel(r'$t^*$')
plt.xlabel(r'$\theta$')
plt.show()
```

```python
def gen_anim(cylcase,doutcase,maxoutcase,fname):
    fig, ax = plt.subplots(figsize=(10,10))

    ax.set_xlim([-0.1, 0.3])
    ax.set_ylim([0, 0.4])

    def init():
        U,W,vort = cylcase.get_fields_at_output(0)
        cs1 = ax.contour(cylcase.XC,cylcase.ZC,vort)

    def update(n): 
        # n = frame counter
        # calculate the positions of the pendulums
        U,W,vort=cylcase.get_fields_at_output(n*doutcase)
        thet,dudz = cylcase.get_dudz(n*doutcase); 
        sep_pt = cylcase.find_zero(thet,dudz)

        thet_0 = (180-sep_pt)*np.pi/180

        ax.cla()
        ax.contourf(cylcase.XC,cylcase.ZC,vort,150,vmin=-1,vmax=.5)
        ax.plot([0, np.cos(thet_0)*0.025],[0,np.sin(thet_0)*0.025],'-k')
        ax.set_xlim([-0.1, 0.3])
        ax.set_ylim([0, 0.4])
        ax.contour(cylcase.XC,cylcase.ZC,vort,levels=[0],colors=['k'])

    anim = animation.FuncAnimation(fig, update, init_func=init, frames=int(maxoutcase/doutcase), blit=True)

    # anim.save can be called in a few different ways, some which might or might not work
    # on different platforms and with different versions of matplotlib and video encoders
    anim.save(fname, fps=5, writer="ffmpeg", codec="libx264")

    plt.close(fig)
```

```python
%%capture --no-stdout
for ii in range(0,NCASES):
    gen_anim(cyl_case[ii],dout[ii],min(40*dout[ii],max_out[ii]),id_str[ii]+".mp4")
    print "Done case " + str(ii+1)
```

```python
def embed_video(ii):
    video = open(id_str[ii]+".mp4", "rb").read()
    video_encoded = video.encode("base64")
    video_tag = '<video controls alt="test" src="data:video/x-m4v;base64,{0}">'.format(video_encoded)
    return video_tag
```
