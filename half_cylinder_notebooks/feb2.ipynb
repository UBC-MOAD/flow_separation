This notebook compares the $Re=50$ and $Re=200$ simulations for various resolutions.

```python
from __future__ import division
import sys
from IPython.display import HTML
sys.path.insert(0, '/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/analysis/')

import cylanalysis as cyl
import numpy as np
from MITgcmutils import rdmds
import matplotlib as mpl
from matplotlib import animation
import matplotlib.pyplot as plt
mpl.rc('text', usetex = True)
mpl.rc('font', family = 'serif')

import matplotlib.pylab as pylab

pylab.rcParams['figure.figsize'] = (10.0, 8.0)

%matplotlib inline
```

```python
%%capture --no-stdout
cyl_case = []

cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re50_Impulse/Nx128Nz128/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re50_Impulse/Nx256Nz128/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re50_Impulse/Nx256Nz256/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re50_Impulse/Nx512Nz256/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re50_Impulse/Nx512Nz512/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re200_Seed/Nx128Nz128/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re200_Seed/Nx256Nz128/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re200_Seed/Nx256Nz256/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re200_Seed/Nx512Nz256/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re200_Seed/Nx512Nz512/'))
cyl_case.append(cyl.CylCase('/ocean/rirwin/Half_Cylinder_MITgcm/Resolution_Dependence/Re2000_Seed/Nx128Nz128/'))
```

```python
NCASES = 11
dout = np.zeros(NCASES)
max_out = np.zeros(NCASES)

id_str = []

#Re = 50
dout[0] = 1000
dout[1] = 1000
dout[2] = 10000
dout[3] = 10000
dout[4] = 10000
max_out[0] = 200000
max_out[1] = 200000
max_out[2] = 2000000
max_out[3] = 200000
max_out[4] = 290000
id_str.append('Re50.128x128')
id_str.append('Re50.256x128')
id_str.append('Re50.256x256')
id_str.append('Re50.512x256')
id_str.append('Re50.512x512')

#Re = 200
dout[5] = 2000
dout[6] = 2000
dout[7] = 2000
dout[8] = 2000
dout[9] = 4000
max_out[5] = 100000
max_out[6] = 100000
max_out[7] = 100000
max_out[8] = 100000
max_out[9] = 200000
id_str.append('Re200.128x128')
id_str.append('Re200.256x128')
id_str.append('Re200.256x256')
id_str.append('Re200.512x256')
id_str.append('Re200.512x512')

#Re = 1000
dout[10] = 1000
max_out[10] = 66000
id_str.append('Re1000.128x128')
```

```python
%%capture --no-stdout
all_sep_pts = []
for jj in range(0,NCASES):
    sep_pts = []
    for ii in range(0,int(max_out[0]/dout[0])+1):
        if ii*dout[jj] <= max_out[jj]:
            thet_tmp,dudz_tmp = cyl_case[jj].get_dudz(dout[jj]*ii)
            sep_pts.append([ii*dout[jj],cyl_case[jj].find_zero(thet_tmp,dudz_tmp)])
    all_sep_pts.append(sep_pts)
    print "Done case " + str(jj) + "."
```

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

plt.figure()

for ii in range(0,NCASES):
    sep_pts = all_sep_pts[ii]
    tt = []
    sep = []
    for jj in sep_pts:
        tt.append((jj[0]/dout[ii])/(0.025/0.01))
        sep.append(jj[1])
    if ii < 5:
        plt.plot(sep,tt,'-o',label=id_str[ii])
    else:
        plt.plot(sep,tt,'-*',label=id_str[ii])

plt.xlim((60,160))
plt.ylim((0,20))
plt.plot([100,180],[0.35,0.35],'-k',label=r'$t_s$')
plt.plot([104.5,104.5],[0,20],'--k',label=r'steady sep.')
plt.legend()
plt.grid()
plt.ylabel(r'$t^*$')
plt.xlabel(r'$\theta$')
plt.show()
```

```python
def gen_anim(cylcase,doutcase,maxoutcase,fname):
    fig, ax = plt.subplots(figsize=(10,5))

    ax.set_xlim([-0.05, 0.15])
    ax.set_ylim([0, 0.075])

    def init():
        U,W,vort = cylcase.get_fields_at_output(0)
        cs1 = ax.contour(cylcase.XC,cylcase.ZC,vort)

    def update(n): 
        # n = frame counter
        # calculate the positions of the pendulums
        U,W,vort=cylcase.get_fields_at_output(n*doutcase)
        thet,dudz = cylcase.get_dudz(n*doutcase); 
        sep_pt = cylcase.find_zero(thet,dudz)

        thet_0 = (180-sep_pt)*np.pi/180

        ax.cla()
        ax.contourf(cylcase.XC,cylcase.ZC,vort,150,vmin=-1,vmax=.5)
        ax.plot([0, np.cos(thet_0)*0.025],[0,np.sin(thet_0)*0.025],'-k')
        ax.set_xlim([-0.05, 0.15])
        ax.set_ylim([0, 0.075])
        ax.contour(cylcase.XC,cylcase.ZC,vort,levels=[0],colors=['k'])

    anim = animation.FuncAnimation(fig, update, init_func=init, frames=int(maxoutcase/doutcase), blit=True)

    # anim.save can be called in a few different ways, some which might or might not work
    # on different platforms and with different versions of matplotlib and video encoders
    anim.save(fname, fps=5, writer="ffmpeg", codec="libx264")

    plt.close(fig)
```

```python
%%capture --no-stdout
for ii in range(0,NCASES):
    gen_anim(cyl_case[ii],dout[ii],min(40*dout[ii],max_out[ii]),id_str[ii]+".mp4")
    print "Done case " + str(ii+1)
```

```python
def embed_video(ii):
    video = open(id_str[ii]+".mp4", "rb").read()
    video_encoded = video.encode("base64")
    video_tag = '<video controls alt="test" src="data:video/x-m4v;base64,{0}">'.format(video_encoded)
    return video_tag
```
