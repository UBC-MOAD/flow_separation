Summary Case 1: Lx=0.6, cvelTimeScale=1
--------------------------------

The maximum anomalies occur at the downstream boundary. Ignoring the rightmost columns of vorticity, the comparison between cases gets much better. The normalized sum of absolute errors is defined as:
\begin{equation}
E_i = \frac{\sum_{x_i} \sum_{z_i} \left| \zeta_i(x_i,z_i) - \zeta_0(x_i,z_i) \right|}{\sum_{x_i}\sum_{z_i} \left| \zeta_0(x_i,z_i) \right|}
\end{equation}
Where $i$ is the case number, $\zeta$ refers to the vorticity. The subscripts in $x$ and $z$ denote that only the $x$ and $z$ points defined for case $i$ are used.

The resulting value of $E_i$ can be interpreted as follows:
- a value of 0 represents exact equivalence between case $i$ and the base case
- a value of 1 represents the difference between a domain with $\zeta_i = 0$ everywhere (i.e. a very poor representation of the base case)

In essence, smaller values denote better agreement.

Using the second reported value of $E_i=0.0014$, it can be concluded that this simulation very closely matches the output of the base case.

How does the normalized sum of absolute errors in vorticity change over time?
===============================================================================

```python
all_saev = []
```

```python
%%capture --no-stdout

for CASE in range(1,4):
    sum_abs_err_vort = []

    for n in range(0,600):
        U,W,vort=cyl_case[CASE].get_fields_at_output(n*dout[CASE])
        U0,W0,vort0=cyl_case[0].get_fields_at_output(n*dout[0])
        
        diff_vort=(vort-vort0[:,0:len(cyl_case[CASE].XC)])
        tot_abs_vort0 = np.sum(np.abs(vort0[:,0:len(cyl_case[CASE].XC)-10]))
        sum_abs_err_vort.append(np.sum(np.abs(diff_vort[:,0:-10]))/tot_abs_vort0)

    all_saev.append(sum_abs_err_vort)
    print "DONE CASE " + str(CASE)
```

Do the sums of absolute errors in vorticity diverge? If $E_i$ is increasing with time for any case, then the solution is inconsistent. If $E_i$ is stable in time, then the solution is relatively consistent.

```python
plt.figure()
plt.plot(all_saev[0])
plt.plot(all_saev[1])
plt.plot(all_saev[2])
plt.legend((id_str[1],id_str[2],id_str[3]))
plt.xlabel('Output \#')
plt.ylabel('$E_i$')
plt.show()
```

It can be seen that the longer domains have errors that are relatively smaller. In the case of $L_x=0.4m$, the error seems to be increasing in time up until 150 outputs. The last to reach steady state is $L_x=0.2m$, around 450 outputs.

This raises the issue of adjustment -- even though the final result of the $L_x=0.2m$ case may be sufficiently close to the desired result, the transient evolution of the system to reach this state is clearly different than that of larger domain cases (similar to Cummins, 2000).
