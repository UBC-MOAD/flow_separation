This script takes in a directory with a flow separation run and generates plots and figures.

```python
from __future__ import division # fix division as floating point division
import numpy as np # get numpy
import matplotlib.pyplot as plt # get plotting
from MITgcmutils import rdmds # get rdmds reading

# plot inline
%matplotlib inline 

# change default size of figures
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12  # that's default image size for this interactive session
```

Set directory and other run-specific parameters here.

```python
_Nx = 256;
_Ny = 1;
_Nz = 64;

_total_depth = 200;

_run_dir = '../../Nx_%d_Nz_%d' % (_Nx,_Nz);

_dt = 20;
_len_day = 86400;
_outputs_per_day = _len_day/_dt;

_out_files = np.arange(0,4*_outputs_per_day,_outputs_per_day);
```

Now load in each field.

```python
_U = np.zeros((_Nz,_Ny,_Nx,len(_out_files)));
_V = np.zeros((_Nz,_Ny,_Nx,len(_out_files)));
_W = np.zeros((_Nz,_Ny,_Nx,len(_out_files)));
_T = np.zeros((_Nz,_Ny,_Nx,len(_out_files)));

_z = np.linspace(-_total_depth,0,_Nz);
_z = _z[::-1];

_x = np.squeeze(rdmds(_run_dir+'/XC'));
_depth = np.squeeze(rdmds(_run_dir+'/Depth'));

for _itr, _out_num in enumerate(_out_files):
    _U[:,:,:,_itr] = rdmds(_run_dir+'/U',_out_num);
    _V[:,:,:,_itr] = rdmds(_run_dir+'/V',_out_num);
    _W[:,:,:,_itr] = rdmds(_run_dir+'/W',_out_num);
    _T[:,:,:,_itr] = rdmds(_run_dir+'/T',_out_num);
  
_U = np.squeeze(_U);
_V = np.squeeze(_V);
_W = np.squeeze(_W);
_T = np.squeeze(_T);
```

```python
def plt_field(_field,_out,_vmin=-.5,_vmax=.5):
    ''' This function plots a specific field at a specific output, and then limits the caxis based on the _vmin/_vmax inputs'''
    plt.pcolor(_x/1000,_z,np.squeeze(_field[:,:,_out]),vmin=_vmin,vmax=_vmax,cmap='Spectral_r'); 
    plt.colorbar();
    plt.plot(_x/1000,-_depth-(_total_depth/2)/_Nz,'-k');
    plt.ylim((-_total_depth,0));
```

Use the previously defined function to output a 2x2 figure of a field at different outputs.

```python
_fig = plt.figure();

_fig.clf();

for _itr,_out in enumerate([0,1,2,3]):
    plt.subplot(2,2,_itr+1);
    plt.title('u vel, day ' + str(_out))
    plt_field(_U,_out,_vmin=0,_vmax=.5);

_fig.subplots_adjust(hspace=0.2,wspace=0.2);
plt.show();
```
